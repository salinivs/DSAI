Q5)Which was the most ordered item? and How many items were ordered?

```python
most = df1.groupby('item_name')
most= most.sum()
most = most.sort_values(['quantity'], ascending=False)
most.head(1)
```

Q6)What was the most ordered item in the choice_description column?

```python
most = df1.groupby('choice_description').sum()
most = most.sort_values(['quantity'], ascending=False)
most.head(1)
```

Q7)Turn the item price into a float

```python
def getprice(strprice):
    strtemp= strprice[1:-1]
    return float(strtemp)
df1['PriceNew']=df1['item_price'].apply(getprice)
df1.head()                                                    # ITEM PRICE IS CONVERT TO FLOAT-COLUMN NAME IS PriceNew
```
